```
import autosklearn
import autosklearn.classification
```

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import warnings 
import sklearn.metrics
```

```
scores = pd.read_csv('combinedall.csv')
```

```
scores.head(5)
```

```
score_columns = [ 'AWSNegative', 'AWSNeutral', 'AWSPositive', 'GCPscore', 'IBMscore','Azurescore']

x_data = np.asarray(scores[score_columns])
y_data = np.asarray(scores['realSentiment'])
```

```
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```
import keras
```

```
y_train_dummies = pd.get_dummies(y_train)
y_test_dummies = pd.get_dummies(y_test)
```

```
autoskl = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=3000)
autoskl.fit(x_train, np.asarray(y_train_dummies))
```

```
predictions = autoskl.predict(x_test)
accuracy_score(y_test_dummies, predictions)
```

```
def convert(x):
    out = []
    for item in x:
        if item == 'positive':
            out.append(2)
        elif item == 'neutral':
            out.append(1)
        else:
            out.append(0)
    return out
    
y_test_num = convert(y_test)
    
confusion_matrix(y_test_num, predictions.argmax(axis=1))
```
